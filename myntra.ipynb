```python
import os           # help to interact with the operating system
import time         # use when there is delay while scrapping
import requests     # As we dealing webpages , we required request
from bs4 import BeautifulSoup as bs     # use to pass the html content
from selenium import webdriver          # seleium use to interact with the browser for dynamic content or interact with component like java and all. Using webdriver we can automate the webpages.
from selenium.webdriver.common.by import By     # This told how selenium locate the web element
```

```python
product = 'anouk kurta'    #Define the path to chrome webdriver executable


# Initializing the chrome webdriver
driver = webdriver.Chrome()

# Navigate the URL
driver.get('https://www.myntra.com/anouk-kurta?rawQuery=' +product)


# Once the page is loaded, you can extract the page text
myntra_text = driver.page_source

myntra_html = bs(myntra_text,"html.parser")
```

```python
myntra_html
```

```python
pclass = myntra_html.find_all('ul',{'class':'results-base'})
```

```python
pclass[0]
```

```python
for i in pclass:
    href = i.find_all('a',href=True)
    t = (href[0]['href'])

product_link = 'https://www.myntra.com/'+t
product_link
```

```python
driver.get(product_link)
proRes = driver.page_source
```

```python
proRes_html = bs(proRes,'html.parser')
proRes_html
```

```python
title_h = proRes_html.find_all('title')
```

```python
Reviews = proRes_html.find('a',{'class':'detailed-reviews-allReviews'})
```

```python
Reviews
```

```python
t2 = Reviews['href']
Reviews_link = 'https://www.myntra.com'+t2
Reviews_link
```

```python
driver.get(Reviews_link)
review_page = driver.page_source
review_page
```

```python
driver.quit()
```

```python
review_html = bs(review_page,'html.parser')
```

```python
review_html
```

```python
review = review_html.find_all('div',{'class':'detailed-reviews-userReviewsContainer'})
```

```python
review
```

```python
for i in review:
    user_rating = i.findAll('div',{'class':'user-review-main user-review-showRating'})
    user_comment = i.findAll('div',{'class':'user-review-reviewTextWrapper'})
    user_name = i.findAll('div',{'class':'user-review-left'})
```

```python
user_rating
user_comment
user_name
```

```python
user_rating
```

```python
for i in range(len(user_rating)):
    rating = user_rating[i].find('span',{'class':'user-review-starRating'}).get_text().strip()
    comment = user_comment[i].text
    name = user_name[i].find('span').text
    date = user_name[i].find_all('span')[1].text
    print(f'Date:{date},Rating:{rating},Name:{name},Comment:{comment}')
```
